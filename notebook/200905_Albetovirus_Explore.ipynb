## Introduction

In the `pmeta5` / `protref5` based search RCE included plant sattelite viruses: ( Accessions ? )

These hit possibly hundreds of related sequences in a few metagenome libraries. A large amount of these contigs are fragments ~300 nt of essentially the capsid alone but many are possibly full length genomes of these viral-like agents.

Distinguishing complete genome vs. partial genome is a little bit tricky, and so far we only know the capsid protein on these elements, expanding this to include possibly conserved RNA elements will be useful for classifying sequences as potential viruses vs. contigs or something else.

## Materials and Methods


````bash
# download materials from rce
cd $WORK
wget https://serratus-public.s3.amazonaws.com/satellite/for_artem.tar.gz
tar -xvf for_artem.tar.gz

ls -alh

rm for_artem.tar.gz
````

````bash
# Copy 97% clustered sequences
cd $WORK
cp combined/all_Albetovirus.id97.fna ./
````

````bash
# Rename headers to be consistent

# >ERR3047578.NODE_1_length_1197_cluster_1_candidate_1_domains_2 
# -->
# >ERR3047578_1_1197nt
cd $WORK

sed 's/NODE_//g' all_Albetovirus.id97.fna \
  | sed 's/_length_.*//g' - \
  > albetovirus.fa
  
samtools faidx albetovirus.fa
````

````bash
# EMBOSS 6.6.0
NAME='albetovirus'

getorf -sequence $NAME.fa \
  -find 1 \
  -minsize 200 \
  -outseq $NAME.orf.fa
````

````bash
# Extract sequences greater than 700 nt (potential genomes)
seqkit sort -rl -w 0 albetovirus.fa > albetovirus.750.fa
samtools faidx albetovirus.750.fa
````

````bash
# EMBOSS 6.6.0
NAME='albetovirus.750'

getorf -sequence $NAME.fa \
  -find 1 \
  -minsize 200 \
  -outseq $NAME.orf.fa
````

````bash
muscle -in albetovirus.750.orf.fa -out albetovirus.750.muscle.fa
````
